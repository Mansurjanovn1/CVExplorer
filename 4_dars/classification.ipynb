```python
import os
from skimage.io import imread
from skimage.transform import resize
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dir = "/content/drive/MyDrive/Colab Notebooks/dataset"
classes = ['cats','dogs']
```

```python
data = []
labels = []
```

```python
classes
```

```python
for class_idx,clas in enumerate(classes):
  for file in os.listdir(os.path.join(dir,clas)):
    img_path = os.path.join(dir,clas,file)
    img = cv2.imread(img_path)
    img = resize(img,(20,20))
    data.append(img.flatten())
    labels.append(class_idx)
data = np.array(data)
labels = np.array(labels)
```

```python
data.shape
```

```python
labels
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,labels,test_size=0.3,stratify =labels,shuffle= True )
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
classifier = SVC()
```

```python
params = [{'gamma' : [0.01,0.001,0.0001],'C' : [1,10,100,1000]}]
```

```python
grid_search = GridSearchCV(classifier,params)
```

```python
 grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_estimator = grid_search.best_estimator_
y_pred = best_estimator.predict(x_test )
```

```python
y_pred
```

```python
score = accuracy_score(y_test,y_pred)
```
