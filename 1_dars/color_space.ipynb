```python
  #Kutubxonalar
import cv2
import matplotlib.pyplot as plt

image_path = "/Users/mansurjonovnurmuhammad/Desktop/color spce/motorbike_0034.jpg" #Rasmni yuklash
image = cv2.imread(image_path) #Rasmni o'qish

if image is None:
    print("Error:") # Agar rasm yoq bolsa "Error" chiqar
else: # Yoki
    rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB) #BGR(Blue-Green-Red) dan RGB(Red-Green-Blue) ga otkazish
    hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV) #BGR(Blue-Green-Red) dan HSV(Hue-Sturation-Value) ga otkazish
    lab_image = cv2.cvtColor(image,cv2.COLOR_BGR2LAB) #BGR(Blue-Green-Red) dan LAB(Ligtnes-a-b) ga otkazish
    greyscale_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #BGR(Blue-Green-Red) dan GRAY(kulrang) ga otkazish
    
    # Rasmlarni korsatish 
    fig,axes = plt.subplots(1,4,figsize=(20,5)) #Matplotlib yordamida 1 qator va 4 ustunlik grafik yaratadi, har biri 20x5

    axes[0].imshow(rgb_image)       #(rgb_image) nomli rasmni chiqarish
    axes[0].set_title("RGB image")  #(RGB image) deb nom berish
    axes[0].axis("off")             #Rasmdagi oq belgilarni yoqotish

    axes[1].imshow(hsv_image)       #(hsv_image) nomli rasmni chiqarish
    axes[1].set_title("HSV image")  #(HSV image) deb nom berish
    axes[1].axis("off")             #Rasmdagi oq belgilarni yoqotish

    axes[2].imshow(lab_image)       #(lab_image) nomli rasmni chiqarish
    axes[2].set_title("LAB image")  #(LAB image) deb nom berish
    axes[2].axis("off")             #Rasmdagi oq belgilarni yoqotish

    axes[3].imshow(greyscale_image, cmap="gray")  #(greyscale_image) nomli rasmni chiqarish
    axes[3].set_title("Greyscale image")  #(Greyscale image) deb nom berish
    axes[3].axis("off")                   #Rasmdagi oq belgilarni yoqotish

    plt.show   #Korsatish



```
